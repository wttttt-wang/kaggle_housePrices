```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
```

```python
# read data from csv
train_data = pd.read_csv('../input/train.csv')
test_data = pd.read_csv('../input/test.csv')
```

```python
train_data.head(2)
```

```python
test_data.head(2)
```

```python
all_data = pd.concat([train_data,test_data])
all_data.info()
```

```python
# pandas took all of the numerical columns and calculated the mean, std, minimum and maximum value
all_data.describe()
```

```python
# an example of scatter
all_data.plot(kind='scatter',x='LotFrontage',y='LotArea')
plt.show()
```

```python
# an example of pairplot, scatter&hist
sns.pairplot(all_data.fillna(0).loc[:,'LotFrontage':'Street'],size=3)
plt.show()
```

```python
all_data.fillna(0).loc[:,'LotFrontage':'Street'].head(1)
```

```python
# showing SalePrice of training data, it's exact what we need to predict for testing data
train_data['SalePrice'].hist()
plt.show()
```

```python
# discretization of SalePrice
# bin: If bins is an int, it defines the number of equal-width bins in the range of x
train_data['discrete_price'] = pd.cut(train_data['SalePrice'],bins=5)
```

```python
train_data['discrete_price'].head(2)
```

```python
sns.boxplot(x='discrete_price',y='LotArea',data=train_data)
sns.stripplot(x='discrete_price',y='LotArea',data=train_data,jitter=True,edgecolor='gray')
plt.show()
```

```python
sns.boxplot(x='discrete_price',y='LotArea',data=train_data)
plt.ylim(0,50000)
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = (12.0, 6.0)
prices = pd.DataFrame({"price":train_data["SalePrice"], "log(price + 1)":np.log1p(train_data["SalePrice"])})
prices.hist()
plt.show()
```
