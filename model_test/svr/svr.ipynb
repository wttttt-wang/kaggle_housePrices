```python
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
```

```python
# read data from csv
def read_data():
    train = pd.read_csv('../input/train_features')
    test = pd.read_csv('../input/test_features')
    label = pd.read_csv('../input/train_label',header=None)
    return train,test,label
```

```python
clf = svm.LinearSVR()
```

```python
X_train,X_test,y = read_data()
from sklearn.model_selection import cross_val_score
print(cross_val_score(clf, X_train, y))
```

```python
X_train,X_test,y = read_data()
from sklearn.model_selection import cross_val_score
print(cross_val_score(clf, X_train, np.array(y),cv=5))
```

```python
X_train,X_test,y = read_data()
from sklearn.model_selection import cross_val_score
print(cross_val_score(clf, X_train, np.array(y),cv=5,scoring='neg_mean_squared_error'))
```

```python
from sklearn.linear_model import Lasso
model_lasso = Lasso()
print(cross_val_score(model_lasso, X_train, np.array(y),cv=5,scoring='neg_mean_squared_error'))
```

```python
import xgboost as xgb
dtrain = xgb.DMatrix(X_train, label = y)
#params = {"max_depth":2, "eta":0.1}
model_xgb = xgb.XGBRegressor(n_estimators=360, max_depth=2, learning_rate=0.1)
print(cross_val_score(model_xgb, X_train, np.array(y),cv=5,scoring='neg_mean_squared_error'))
```
